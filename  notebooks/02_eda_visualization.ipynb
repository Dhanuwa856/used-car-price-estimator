```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load the dataset
file_path = '../data/processed/used_cars_dataset_cleaned.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# Distribution of Selling Price
plt.figure(figsize=(15, 6))

sns.histplot(df['AskPrice'], bins=100, kde=True,log_scale=True)
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# Fuel_Type to Price
plt.figure(figsize=(15, 6))
sns.boxplot(x='FuelType', y='AskPrice', data=df)
plt.title('Fuel Type vs Selling Price')
plt.xlabel('Fuel Type')
plt.ylabel('Selling Price')
plt.show()
```

```python
# saller type to price
plt.figure(figsize=(15, 6))
sns.boxplot(x='Owner', y='AskPrice', data=df)
plt.title('Owner vs Selling Price')
plt.xlabel('Owner')
plt.ylabel('Selling Price')
plt.show()
```

```python
# transmission to price
plt.figure(figsize=(15, 6))
sns.boxplot(x='Transmission', y='AskPrice', data=df)
plt.title('Transmission vs Selling Price')
plt.xlabel('Transmission')
plt.ylabel('Selling Price')
plt.show()
```

```python
# car age to price
plt.figure(figsize=(15, 6))
sns.boxplot(x='Car_Age', y='AskPrice', data=df)
plt.title('Car Age vs Selling Price')
plt.xlabel('Car Age')
plt.ylabel('Selling Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# s=10 මගින් තිත් වල ප්‍රමාණය කුඩා කරයි
sns.scatterplot(x='Car_Age', y='AskPrice', data=df, s=10, alpha=0.5)
plt.title('Selling Price vs Kms Driven (Smaller Dots)')
plt.show()
```
