```python
import pandas as pd
import numpy as np
```

```python
file_path = '../data/ raw/used_cars_dataset.csv'
```

```python
df = pd.read_csv(file_path)
df.head()
```

```python
print("Number of rows: ", df.shape[0])
print("Number of columns: ", df.shape[1])
```

```python
print(df.info())
```

```python
print("check null values")
print(df.isnull().sum())
```

```python
print("check duplicates")
print(df.duplicated().sum())
```

```python
df_cleaned = df.drop_duplicates()
print("Number of rows after removing duplicates: ", df_cleaned.shape[0])
```

```python
from datetime import datetime
now = datetime.now()
df_cleaned['Car_Age'] = now.year - df_cleaned['Year']
```

```python
df_cleaned.head()
```

```python
df_cleaned = df_cleaned.drop(columns=['Year'], inplace=False)
```

```python
df_cleaned.head()
```

```python
df_cleaned.to_csv('../data/processed/used_cars_dataset_cleaned.csv', index=False)
```
